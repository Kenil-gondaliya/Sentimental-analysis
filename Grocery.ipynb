```python
import re 
from textblob import TextBlob 
import emoji  


import re 
from textblob import TextBlob 
import emoji  
def clean_tweet(tweet): 
    tweet = re.sub(r'@[A-Za-z0-9]+', '', str(tweet)) # remove @mentions
    tweet = re.sub(r'#', '',  str(tweet)) # remove the '#' symbol
    tweet = re.sub(r'RT[\s]+', '',  str(tweet)) # remove RT
    tweet = re.sub(r'https?\/\/S+', '',  str(tweet)) # remove the hyperlink
    tweet = re.sub(r'http\S+', '',  str(tweet)) # remove the hyperlink
    tweet = re.sub(r'www\S+', '',  str(tweet)) # remove the www
    tweet = re.sub(r'twitter+', '',  str(tweet)) # remove the twiiter
    tweet = re.sub(r'pic+', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'com', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'africa', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'innovation', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'covid-19', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'coronavirus', '',  str(tweet)) # remove the pic
    tweet = re.sub(r'technology', '',  str(tweet)) # remove the pic

    return tweet


def remove_emoji(tweet):
    return emoji.get_emoji_regexp().sub(u'', tweet)

```

```python
comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df2.cleaned_text: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 1000, height = 800, 
                background_color='white', colormap='Set2', 
                collocations=False, 
                stopwords = stopwords, 
                min_font_size = 12).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (10, 10), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
# get functionality of subjectivity and polarity 
def getSubjectivity(text):
    return TextBlob( str(text)).sentiment.subjectivity

def getPolarity(text):
    return TextBlob( str(text)).sentiment.polarity
```

```python
df2.dropna(subset=['cleaned_text'], inplace = True)
df2.reset_index(drop=True, inplace=True)
```

```python
df2['Subjectivity'] = df2['cleaned_text'].apply(getSubjectivity)
df2['Polarity'] = df2['cleaned_text'].apply(getPolarity)
df2.head()
```

```python
def get_Polarity_Analysis(score):
    if score < 0:
      return 'Negative'
    elif score == 0:
      return 'Neutral'
    else:
      return 'Positive'
def get_Subjectivity_Analysis(score):
    if score >  0:
      return 'Opinion'
    else:
      return 'Fact'

df2['Analysis_Polarity'] = df2['Polarity'].apply(get_Polarity_Analysis)

df2['Analysis_Subjectivity'] = df2['Subjectivity'].apply(get_Subjectivity_Analysis)

# Show the dataframe
df2.head(20)

```

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b')
                                #max_df = 0.5, 
                                #min_df = 10)
                               # max_df = 10, 
                               # min_df = .5)
dtm_tf = tf_vectorizer.fit_transform(df2['cleaned_text'].values.astype('U'))
print(dtm_tf.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df2 contains your data with columns Polarity, Subjectivity, and Topic

# Set figure size
plt.figure(figsize=(40, 25))

# Create lmplot with hue='Topic'
g = sns.lmplot(x="Polarity", y="Subjectivity", hue='Topic', data=df2, fit_reg=False, palette="GnBu_d", legend=True, col='Topic', legend_out=True)

# Add counts as text annotations
def add_counts(data, **kwargs):
    if not data.empty:
        topic = data['Topic'].iloc[0]  # Get the topic for the current facet (assuming only one topic per facet)
        counts = data.groupby(['Polarity', 'Subjectivity']).size().reset_index(name='Counts')
        for _, row in counts.iterrows():
            plt.text(row['Polarity'], row['Subjectivity'], s=row['Counts'], 
                     horizontalalignment='center', verticalalignment='center', 
                     fontsize=10, color='black')

# Iterate over each facet and add annotations
for ax in g.axes.flat:
    data = df2[df2['Topic'] == ax.get_title()]  # Get data for the current facet
    add_counts(data)

# Adjust layout
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Polarity vs. Subjectivity by Topic')

# Show the plot
plt.show()
```
